```python
import pandas as pd
import numpy as np
```

```python
pd.options.mode.chained_assignment = None 
```

```python
file = '/Users/lmp/Documents/Programming/CNAM/SIG/Fracture_numerique/Data/mci/Agregation_MCI_par_Commune_commune-techno-beta.csv'
```

```python
mci_communes = pd.read_csv(file, delimiter=";", skiprows=1)
```

```python
mci_communes = mci_communes.convert_dtypes()
```

```python
mci_communes_ARA = mci_communes[mci_communes["code_reg"]==84]
```

```python
clean = mci_communes_ARA.groupby('code_insee')['date'].agg(lambda x: x.sort_values().take([-1]))
```

```python
len(clean)
```

```python
mci_communes_ARA = pd.merge(mci_communes_ARA, clean, how='inner', on=['code_insee', 'date'])
```

```python
mci_communes_ARA.sort_values(by='nom_com')
```

## Richesse de l'offre : nombre de MCI par commune

```python
mcis = ['log_cu', 'log_coax', 'log_ftth', 'log_4gf', 'log_thdr', 'log_hdr', 'log_sat']
```

```python
mci_communes_ARA['nbr_mci'] = 0
```

```python
for index, row in mci_communes_ARA.iterrows():
    for mci in mcis :
        if row[mci] > 0 :
            mci_communes_ARA.loc[index, 'nbr_mci'] += 1
```

## Taux d'accessibilité par offre par commune

```python
mci_communes_ARA['taux_log_cu'] = 0
mci_communes_ARA['taux_log_coax'] = 0
mci_communes_ARA['taux_log_ftth'] = 0
mci_communes_ARA['taux_log_4gf'] = 0
mci_communes_ARA['taux_log_thdr'] = 0
mci_communes_ARA['taux_log_hdr'] = 0
mci_communes_ARA['taux_log_sat'] = 0
```

```python
for index, row in mci_communes_ARA.iterrows():
    for mci in mcis :
        mci_communes_ARA.loc[index, f'taux_{mci}'] = mci_communes_ARA.loc[index, mci] / mci_communes_ARA.loc[index, 'nbr']
```

```python
mci_communes_ARA.groupby('code_insee').count()
```

## Taux d'accessibilité global par commune

```python
for index, row in mci_communes_ARA.iterrows():
    column_true = []
    for mci in mcis :
        if mci_communes_ARA.loc[index, f'taux_{mci}'] > 0 :
            column_true.append(f'taux_{mci}')
            mci_communes_ARA.loc[index, 'taux_global'] = mci_communes_ARA.loc[index, column_true].mean()
```

```python
mci_communes_ARA.to_csv('mci_communes_ARA.csv')
```
